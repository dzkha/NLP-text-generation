```
#Open text file and identify all the adjectives in it

import nltk
with open('gh.txt', 'r') as file:
    gh = file.read().replace('\n', '')

text = nltk.word_tokenize(gh)
tagged = nltk.pos_tag(text)
adjectives = set()
for i in tagged:
  if i[1] == 'JJ':
    adjectives.add(i[0])

adjectives
```
